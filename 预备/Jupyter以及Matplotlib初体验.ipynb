# 使用Jupyter Notebook

## 常用快捷键
**命令模式下(ESC)(蓝色)**
- `A` 向上添加一个单元格
- `B` 向下添加一个单元格
- `DD` 删除单元格
- `M` 将单元格转换为Markdown格式
- `Y` 将单元格转化为代码模式
  
**编辑模式下(按下Enter)(绿色)**
- `Enter + Shift` 执行单元格并切换到下一单元格
- `Enter + Ctrl` 执行单元格

## 关于撤销删除操作
- 删除了整个单元格
    在命令模式下按`Z`恢复单元格 **注意 仅在web端上的notebook上可行，vscode上无法恢复**
- 查找历史代码
    在单元格内输入 `history` 可以查看所有执行了的代码

# 案例： 绘制温度折线图

```python
import random
x= range(60)
y=[random.uniform(10,15) for i in x] 
y_2=[random.uniform(12,20) for i in x]
# y=[random.uniform(10,15) for i in x if i%5==0] 是列表的推导式子 等价于以下写法
# y=[]
# for i in x:
#     if(i%5==0):
#         y.append(random.uniform(10,15))


plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y,label="y1")
plt.plot(x,y_2,linestyle="--",color="r",label="y2")
# 导入中文字体
plt.rcParams['font.sans-serif']=['SimHei']#黑体
#生成xy轴标签
x_ticks_label=["{}分".format(i) for i in x]
y_ticks=range(10,20)
#这里函数的第一个参数必须是数字，如果不是数字要进行替换，不能直接传字符串
plt.xticks(x[::5],x_ticks_label[::5],fontsize=20)
plt.yticks(y_ticks[::1],fontsize=20)
#添加网格
plt.grid(True,linestyle="--",alpha=0.5)
# 添加描述信息
plt.xlabel("时间",fontsize=20)
plt.ylabel("温度",fontsize=20)
plt.title("温度随时间变化折线",fontsize=20)
#添加图例
plt.legend(loc="best")# 添加图例前 必须要在plot上写明label
plt.show()
```

### plot 和legend中的一些参数
| 颜色字符color= | 风格字符linestyle= |
|:----:|:----:|
|r 红色|- 实线|
|g 绿色|-- 虚线|
|b 蓝色|-. 点划线|
|w 白色|: 点虚线|
|c 青色|' ' 留空|
|m 洋红|
|y 黄色|
|k 黑色|

# 案例： 绘制数学函数图形

```python
import numpy as np
x= np.linspace(-10,10.,1000)
y= np.sin(x)
# 修正显示不了负号的情况
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(20,8),dpi=100)
plt.plot(x,y)
plt.show()
```
