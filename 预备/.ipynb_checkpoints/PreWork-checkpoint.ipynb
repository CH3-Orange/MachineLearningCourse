# 使用Jupyter Notebook

## 常用快捷键
**命令模式下(ESC)(蓝色)**
- `A` 向上添加一个单元格
- `B` 向下添加一个单元格
- `DD` 删除单元格
- `M` 将单元格转换为Markdown格式
- `Y` 将单元格转化为代码模式
  
**编辑模式下(按下Enter)(绿色)**
- `Enter + Shift` 执行单元格并切换到下一单元格
- `Enter + Ctrl` 执行单元格

## 关于撤销删除操作
- 删除了整个单元格
    在命令模式下按`Z`恢复单元格 **注意 仅在web端上的notebook上可行，vscode上无法恢复**
- 查找历史代码
    在单元格内输入 `history` 可以查看所有执行了的代码

# 使用Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# 创建画布
plt.figure()
x=[0,1,2,3,4,5]
y=[5,6,4,3,5,2]
plt.plot(x,y)

# 展示图像
plt.show()
```

```python
# 查看函数的参数
# help(plt.figure)
#?plt.figure
# 末尾加上分号阻止这一语句的执行
```

```python
plt.figure(figsize=(20,8),dpi=100)
x=[0,1,2,3,4,5]
y=[5,6,4,3,5,2]
plt.plot(x,y)
# 图片保存
# 注意要先保存才能show，show会释放对象，show后会保存不到原始数据的图片
# 同时，路径必须是存在的路径，不存在的目录是不会自动生成的
plt.savefig('./pic/test1.jpg')
# 展示图像
plt.show()
```

# 案例： 绘制温度曲线图

```python
import random
x= range(60)
y=[random.uniform(10,15) for i in x] 
# y=[random.uniform(10,15) for i in x if i%5==0] 是列表的推导式子 等价于以下写法
# y=[]
# for i in x:
#     if(i%5==0):
#         y.append(random.uniform(10,15))


plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)
# 导入中文字体
plt.rcParams['font.sans-serif']=['SimHei']#黑体
#生成xy轴标签
x_ticks_label=["{}分".format(i) for i in x]
y_ticks=range(10,15)
#这里函数的第一个参数必须是数字，如果不是数字要进行替换，不能直接传字符串
plt.xticks(x[::5],x_ticks_label[::5])
plt.yticks(y_ticks[::1])
#添加网格
plt.grid(True,linestyle="--",alpha=0.5)
plt.show()
```
